# Choosing a K Value
Let's go ahead and use the elbow method to pick a good K Value!

*Create a for loop that trains various KNN models with different k values, then keep track of the error_rate for each of these models with a list.*

```python
##########################
##########################

## TASK-6 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################


import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

def plot_KNN_error_rate(xdata, ydata):
    error_rate = []
    test_scores = []
    train_scores = []

    ## Split the data into train and test sets in a 70:30 ratio (70% train, 30% test)
    X_train, X_test, y_train, y_test = train_test_split(xdata, ydata, test_size=0.3, random_state=42)

    for i in range(1, 40):
        ## Initialize the KNN classifier with 'i' neighbors
        knn = KNeighborsClassifier(n_neighbors=i)

        ## Fit the KNN model on the training set
        knn.fit(X_train, y_train)

        ## Make predictions on the test set using KNN
        pred_i = knn.predict(X_test)

        error_rate.append(np.mean(pred_i != y_test))  # Calculate error rate
        train_scores.append(knn.score(X_train, y_train))  # Train score
        test_scores.append(knn.score(X_test, y_test))  # Test score

    # Plotting the error rate vs. K value
    plt.figure(figsize=(12, 8))
    plt.plot(range(1, 40), error_rate, color='blue', linestyle='dashed', marker='o',
             markerfacecolor='red', markersize=10)
    plt.title('Error Rate vs. K Value')
    plt.xlabel('K')
    plt.ylabel('Error Rate')
    plt.show()

    # Print max train score and k
    max_train_score = max(train_scores)
    train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
    print('Max train score {} % and k = {}'.format(max_train_score * 100, list(map(lambda x: x + 1, train_scores_ind))))

    # Print max test score and k
    max_test_score = max(test_scores)
    test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
    print('Max test score {} % and k = {}'.format(max_test_score * 100, list(map(lambda x: x + 1, test_scores_ind))))

    return test_scores
```

```python
## TASK-7: Refer to MinMax Scaler provided in scikit-learn.
## Use MinMax scaling on the dataset, and see the performance of KNN on this minmax-scaled dataset.


import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler

def plot_KNN_error_rate_scaled(xdata, ydata):
    error_rate = []
    test_scores = []
    train_scores = []

    ## [Step 1] Apply MinMax scaling to the dataset
    scaler = MinMaxScaler()
    xdata_scaled = scaler.fit_transform(xdata)  # Scale the features to [0, 1] range

    ## [Step 2] Split the scaled data into train and test sets in a 70:30 ratio (70% train, 30% test)
    X_train, X_test, y_train, y_test = train_test_split(xdata_scaled, ydata, test_size=0.3, random_state=42)

    for i in range(1, 40):
        ## Initialize the KNN classifier with 'i' neighbors
        knn = KNeighborsClassifier(n_neighbors=i)

        ## Fit the KNN model on the training set
        knn.fit(X_train, y_train)

        ## Make predictions on the test set using KNN
        pred_i = knn.predict(X_test)

        error_rate.append(np.mean(pred_i != y_test))  # Calculate error rate
        train_scores.append(knn.score(X_train, y_train))  # Train score
        test_scores.append(knn.score(X_test, y_test))  # Test score

    # Plotting the error rate vs. K value
    plt.figure(figsize=(12, 8))
    plt.plot(range(1, 40), error_rate, color='blue', linestyle='dashed', marker='o',
             markerfacecolor='red', markersize=10)
    plt.title('Error Rate vs. K Value after MinMax Scaling')
    plt.xlabel('K')
    plt.ylabel('Error Rate')
    plt.show()

    # Print max train score and k
    max_train_score = max(train_scores)
    train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
    print('Max train score {} % and k = {}'.format(max_train_score * 100, list(map(lambda x: x + 1, train_scores_ind))))

    # Print max test score and k
    max_test_score = max(test_scores)
    test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
    print('Max test score {} % and k = {}'.format(max_test_score * 100, list(map(lambda x: x + 1, test_scores_ind))))

    return test_scores
```

```python
## TASK-8:
## Plot the voronoi diagram for the TASK-7


import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

def plot_voronoi_diagram(xdata, ydata, k=3):
    # Step 1: Apply MinMax scaling to the dataset
    scaler = MinMaxScaler()
    xdata_scaled = scaler.fit_transform(xdata)  # Scale the features to [0, 1] range

    # Step 2: Split the scaled data into train and test sets in a 70:30 ratio (70% train, 30% test)
    X_train, X_test, y_train, y_test = train_test_split(xdata_scaled, ydata, test_size=0.3, random_state=42)

    # Step 3: Initialize the KNN classifier with 'k' neighbors
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)  # Fit the KNN model on the training set

    # Step 4: Generate a mesh grid over the feature space
    h = .02  # Step size in the mesh
    x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
    y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1

    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    # Step 5: Make predictions for each point on the mesh grid
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)  # Reshape predictions back into a grid

    # Step 6: Plot the Voronoi diagram
    plt.figure(figsize=(10, 8))
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.coolwarm, s=50, edgecolors='k', marker='o', label='Training points')
    plt.title(f'Voronoi Diagram for KNN (k={k})')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.show()

# Example usage:
# plot_voronoi_diagram(xdata, ydata, k=3)  # Replace xdata, ydata with your dataset
```

```python
## TASK-9: Use K-Fold cross validation on all the above classification experiments and present an analysis of the results you obtain.

import numpy as np
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

def k_fold_cross_validation(xdata, ydata, k=5):
    # Step 1: Apply MinMax scaling to the dataset
    scaler = MinMaxScaler()
    xdata_scaled = scaler.fit_transform(xdata)  # Scale the features to [0, 1] range

    # Initialize KFold cross-validation with 'k' folds
    kf = KFold(n_splits=k, shuffle=True, random_state=42)

    # Step 2: KNN on the original dataset (without scaling)
    knn = KNeighborsClassifier(n_neighbors=3)  # You can adjust k as needed
    original_scores = cross_val_score(knn, xdata, ydata, cv=kf, scoring='accuracy')

    # Step 3: KNN on the scaled dataset (with MinMax scaling)
    scaled_scores = cross_val_score(knn, xdata_scaled, ydata, cv=kf, scoring='accuracy')

    # Step 4: Print the results for both models
    print("KNN on Original Data (without scaling):")
    print(f"Accuracy Scores for each fold: {original_scores}")
    print(f"Mean Accuracy: {np.mean(original_scores)}")
    print(f"Standard Deviation of Accuracy: {np.std(original_scores)}")
    print()

    print("KNN on Scaled Data (MinMax Scaling):")
    print(f"Accuracy Scores for each fold: {scaled_scores}")
    print(f"Mean Accuracy: {np.mean(scaled_scores)}")
    print(f"Standard Deviation of Accuracy: {np.std(scaled_scores)}")

    # Step 5: Visualize the results
    plt.figure(figsize=(10, 6))
    plt.boxplot([original_scores, scaled_scores], labels=['Original Data', 'Scaled Data'])
    plt.title('K-Fold Cross Validation Results: KNN')
    plt.ylabel('Accuracy')
    plt.show()

# Example usage:
# k_fold_cross_validation(xdata, ydata, k=5)  # Replace xdata, ydata with your dataset
```
